```python
from tkinter import *
import tkinter as tk
import tkinter.font
import tkinter.simpledialog
import tkinter.messagebox
import csv
import openpyxl,xlrd
from openpyxl import Workbook
from openpyxl import load_workbook
import pathlib  
root=Tk()
name=StringVar()
age=IntVar()
j=IntVar()
k=IntVar()
i=IntVar() 
bioage=IntVar()
x=IntVar()
x.set(5)
def submit():
    name1=name.get()
    age1=age.get()
    z1=j.get()
    z2=i.get()
    z3=k.get()
    z4=0
    if (name1==""):
        print("error")
        tkinter.messagebox.showerror("error","mistake in data entry")
        name1.set("")
        age1.set("")
    elif (age1==""):
        print("error")
        tkinter.messagebox.showerror("error","please enter your age")
    
    if (z1==1):
        if(z2==1):
            if(z3==1):
                z4=age1*1.19
            if(z3==2):
                 z4=age1*1.12
        if(z2==2):
            if(z3==1):
                z4=age1*1.16
            if(z3==2):
                 z4=age1*1.07
    if (z1==2):
        if(z2==1):
            if(z3==1):
                z4=age1*1.14
            if(z3==2):
                 z4=age1*1.10
        if(z2==2):
            if(z3==1):
                z4=age1*1.09
            if(z3==2):
                 z4=age1*1.001
    x.set(z4)
    e5=Label(root,textvariable=x,font=(15),bg="yellow").grid(row=8,column=2)
    
    with open('newdata.csv','a')as csvfile:
            writer=csv.writer(csvfile)
            writer.writerow([name1,age1,z1,z2,z3,z4])
            csvfile.close()
        
         

root.geometry("250x170")
root.title("Motivation world")
root.configure(bg="#a8dadc")
l = Label(root, text = "This is a platform that reflects your health status",font=('Arial',24,'bold')).grid(row=0,columnspan=4,padx=20,pady=10)
l3=Label(root,text="Enter Your name",font=(15),bg="yellow").grid(row=1,column=0)
e3=Entry(root,textvariable=name,width=30,font=(15),bg="yellow").grid(row=1,column=1)
l1=Label(root,text="Enter age",font=(15),bg="yellow").grid(row=2,column=0)
e1=Entry(root,textvariable=age,width=30,font=(15),bg="yellow").grid(row=2,column=1)
l2=Label(root,text="Enter Height",font=(15),bg="yellow").grid(row=3,column=0)
e2=Entry(root,width=30,font=(15),bg="yellow").grid(row=3,column=1)
l3=Label(root,text="Do you smoke?",font=(15),bg="pink").grid(row=4,column=0)
rad1=Radiobutton(root,text="Yes",font=(15),value=1,variable=j,bg="pink").grid(row=4,column=1)
rad2=Radiobutton(root,text="No",value=2,font=(15),variable=j,bg="pink").grid(row=4,column=2)
l4=tkinter.Label(root,text="Do you have COPD?",font=(15),bg="pink").grid(row=5,column=0)
rad3=Radiobutton(root,text="Yes",font=(15),value=1,variable=i,bg="pink").grid(row=5,column=1)
rad4=Radiobutton(root,text="No",value=2,font=(15),variable=i,bg="pink").grid(row=5,column=2)
l4=Label(root,text="Do you have asthma?",font=(15),bg="pink").grid(row=6,column=0)
rad3=Radiobutton(root,text="Yes",font=(15),value=1,variable=k,bg="pink").grid(row=6,column=1)
rad4=Radiobutton(root,text="No",value=2,font=(15),variable=k,bg="pink").grid(row=6,column=2)
bt=Button(root,text="submit",font=(15),bg="pink",command=submit).grid(row=7,column=0) 
l5=Label(root,text="Your biological age is",font=(15),bg="red").grid(row=8,column=0)

b1 = Button(root, text = "For consultants",padx=20,bg="#f1faee").grid(row=9,column=2)
T = Text(root, height = 5, width = 60, bg="#f1faee",fg="black", relief=SUNKEN)
Fact = """Smoking is one of the most dangerous habit that can deteriorate your
lung conditions and along with other lung diseases can increase your
biological age to an undesirous amount."""
T.insert(tk.END, Fact)
T.config(font =("Helvetica", 18,'normal'),bg="#48cae4")
root.mainloop()

```
